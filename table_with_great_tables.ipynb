```python
import pandas as pd
import numpy as np

np.random.seed(42)

regions = ["North", "South", "East", "West"]
products = ["A", "B", "C"]

data = []
for r in regions:
    for p in products:
        revenue = np.random.randint(50000, 200000)
        profit = revenue * np.random.uniform(0.1, 0.4)
        margin = profit / revenue

        data.append({
            "region": r,
            "product": p,
            "revenue": revenue,
            "profit": profit,
            "margin": margin
        })

df = pd.DataFrame(data)
```

```python
from great_tables import GT

gt_tbl = (
    GT(df)

    # Title + subtitle
    .tab_header(
        title="📊 Regional Product Performance",
        subtitle="Synthetic Sales Data (2026)"
    )

    # Column labels
    .cols_label(
        region="Region",
        product="Product",
        revenue="Revenue",
        profit="Profit",
        margin="Profit Margin"
    )

    # Format numbers
    .fmt_currency(
        columns=["revenue", "profit"],
        currency="USD"
    )
    .fmt_percent(
        columns="margin",
        decimals=1
    )

    # Add color scales
    .data_color(
        columns="revenue",
        palette="Greens"
    )
    .data_color(
        columns="margin",
        palette="Blues"
    )

    # Align text nicely
    .cols_align(
        align="center",
        columns=["region", "product"]
    )

    # Group columns visually
    .tab_spanner(
        label="Financials",
        columns=["revenue", "profit", "margin"]
    )

    # Add a source note
    .tab_source_note(
        source_note="Source: Synthetic dataset for demonstration"
    )
)

gt_tbl
```

```python
html = gt_tbl.as_raw_html()

with open("table.html", "w") as f:
    f.write(html)
```
