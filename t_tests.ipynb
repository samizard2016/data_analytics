```python
import numpy as np

from scipy.stats import ttest_1samp, ttest_ind, ttest_rel

# Create data
data=np.random.normal(60,25,100)

# Find mean
mean_value = np.mean(data)

print(f"Mean: {mean_value}")
# Perform one-sample t-test
t_test_value, p_value = ttest_1samp(data, 60)
print(f"t-test Value: {t_test_value} with p-value: {p_value}")

# 0.05 or 5% is significance level or alpha.
comment = "has significant difference from 60" if p_value < 0.05 else "has no significant difference from 60"
print(f"Mean value {mean_value:.0f} {comment}")

# Create data for two sample t-test
data1 = np.random.normal(60,25,100)
data2 = np.random.normal(60,15,100)

tstat, p = ttest_ind(data1, data2)
comment = "is significant" if p_value < 0.05 else "is not significant"
print(f"\nMean1: {data1.mean():.1f} Mean2: {data2.mean():.1f}\nt-test Value: {tstat:.3f} with p-value: {p} result: the mean difference ({(data1.mean()-data2.mean()):.2f}) {comment}")

print("\nPair t-test: if the samples are correlated - eg. sample person has scored in two attempts")

tstat, p = ttest_rel(data1, data2)
comment = "is significant" if p_value < 0.05 else "is not significant"
print(f"Mean1: {data1.mean():.1f} Mean2: {data2.mean():.1f}\nt-test Value: {tstat:.3f} with p-value: {p} result: the mean difference ({(data1.mean()-data2.mean()):.2f}) {comment}")
```

### Estimating Cohen's d for the difference between two sample means

Cohen's d is an effect size measure that quantifies the difference between two means in standard deviation units. It's often used to interpret the practical significance of a statistical result, beyond just the p-value.

The formula for Cohen's d for two independent samples is:

$d = \frac{\bar{X}_1 - \bar{X}_2}{s_p}$

Where:
* $\bar{X}_1$ and $\bar{X}_2$ are the means of the two samples.
* $s_p$ is the pooled standard deviation.

The pooled standard deviation ($s_p$) is calculated as:

$s_p = \sqrt{\frac{(n_1 - 1)s_1^2 + (n_2 - 1)s_2^2}{n_1 + n_2 - 2}}$

Where:
* $n_1$ and $n_2$ are the sizes of the two samples.
* $s_1$ and $s_2$ are the standard deviations of the two samples.

```python
import numpy as np

# Assuming data1 and data2 are based off 20 trials for considering Hedge's g
data1 = np.random.normal(60,25,20)
data2 = np.random.normal(60,15,20)

# Calculate means
mean1 = np.mean(data1)
mean2 = np.mean(data2)

# Calculate standard deviations
std1 = np.std(data1, ddof=1) # ddof=1 for sample standard deviation
std2 = np.std(data2, ddof=1) # ddof=1 for sample standard deviation

# Calculate sample sizes
n1 = len(data1)
n2 = len(data2)

# Calculate pooled standard deviation
pooled_std = np.sqrt(((n1 - 1) * std1**2 + (n2 - 1) * std2**2) / (n1 + n2 - 2))

# Calculate Cohen's d
cohens_d = (mean1 - mean2) / pooled_std

print(f"Mean of data1: {mean1:.2f}")
print(f"Mean of data2: {mean2:.2f}")
print(f"Standard deviation of data1: {std1:.2f}")
print(f"Standard deviation of data2: {std2:.2f}")
print(f"Pooled standard deviation: {pooled_std:.2f}")
print(f"Cohen's d: {cohens_d:.3f}")

# Interpret Cohen's d (common guidelines)
if abs(cohens_d) < 0.2:
    effect_size_comment = "(trivial effect)"
elif abs(cohens_d) < 0.5:
    effect_size_comment = "(small effect)"
elif abs(cohens_d) < 0.8:
    effect_size_comment = "(medium effect)"
else:
    effect_size_comment = "(large effect)"

print(f"The effect size (Cohen's d) of {cohens_d:.3f} indicates a {effect_size_comment} between the two sample means.")
```

### Estimating Hedges' g

Hedges' g is an unbiased estimator of effect size, often preferred over Cohen's d when sample sizes are small (typically $n < 20$). It applies a correction factor to Cohen's d to account for bias.

The formula for Hedges' g is:

$g = d \times J$

Where:
* $d$ is Cohen's d.
* $J$ is the correction factor, calculated as:

$J = 1 - \frac{3}{4(n_1 + n_2 - 2) - 1}$

```python
# Calculate the correction factor J
# n1 and n2 are sample sizes from the previous calculation
# cohens_d is also from the previous calculation

df_total = n1 + n2 - 2  # Total degrees of freedom
correction_factor_j = 1 - (3 / (4 * df_total - 1))

# Calculate Hedges' g
hedges_g = cohens_d * correction_factor_j

print(f"Cohen's d: {cohens_d:.3f}")
print(f"Correction factor J: {correction_factor_j:.3f}")
print(f"Hedges' g: {hedges_g:.3f}")

# Interpret Hedges' g (similar guidelines to Cohen's d)
if abs(hedges_g) < 0.2:
    hedges_effect_size_comment = "(trivial effect)"
elif abs(hedges_g) < 0.5:
    hedges_effect_size_comment = "(small effect)"
elif abs(hedges_g) < 0.8:
    hedges_effect_size_comment = "(medium effect)"
else:
    hedges_effect_size_comment = "(large effect)"

print(f"The adjusted effect size (Hedges' g) of {hedges_g:.3f} indicates a {hedges_effect_size_comment} between the two sample means.")
```
