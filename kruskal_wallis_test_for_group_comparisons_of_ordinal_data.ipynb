```python
from scipy.stats import kruskal
import numpy as np

# Data
x = np.random.normal(35,15,10)
y = np.random.normal(45,20,10)
z = np.random.normal(55,25,10)

# Apply kruskal-wallis test
# The Kruskal-Wallis H-test tests the null hypothesis that the population
# median of all of the groups are equal.  It is a non-parametric version of
# ANOVA.  The test works on 2 or more independent samples, which may have
# different sizes.  Note that rejecting the null hypothesis does not
# indicate which of the groups differs.  Post hoc comparisons between
# groups are required to determine which groups are different.
stat, p = kruskal(x,y,z)

print(f"x,y,z with means {np.mean(x):.2f}, {np.mean(y):.2f}, {np.mean(z):.2f}\n" +
      f" and standard deviations {np.std(x):.2f}, {np.std(y):.2f},{np.std(z):.2f} respectively")
comment = "significant" if p < 0.05 else "not significant"
print(f"Their distribution is statistically {comment} at 95% level with the value of p: {p:.3f}")
```

```python
from scipy.stats import kruskal
import numpy as np

# Data (modified to likely produce a significant result for demonstration)
x = np.random.normal(30, 5, 10) # Group 1
y = np.random.normal(40, 5, 10) # Group 2, mean is different
z = np.random.normal(30, 5, 10) # Group 3, similar to Group 1

# Apply kruskal-wallis test
# The Kruskal-Wallis H-test tests the null hypothesis that the population
# median of all of the groups are equal.  It is a non-parametric version of
# ANOVA.  The test works on 2 or more independent samples, which may have
# different sizes.  Note that rejecting the null hypothesis does not
# indicate which of the groups differs.  Post hoc comparisons between
# groups are required to determine which groups are different.
stat, p = kruskal(x, y, z)

print(f"x,y,z with means {np.mean(x):.2f}, {np.mean(y):.2f}, {np.mean(z):.2f}\n" +
      f" and standard deviations {np.std(x):.2f}, {np.std(y):.2f},{np.std(z):.2f} respectively")
comment = "significant" if p < 0.05 else "not significant"
print(f"Their distribution is statistically {comment} at 95% level with the value of p: {p:.3f}")
```

```python
import scikit_posthocs as sp
import pandas as pd

# Combine data into a single DataFrame for scikit_posthocs
data = pd.DataFrame({
    'value': np.concatenate([x, y, z]),
    'group': ['x']*len(x) + ['y']*len(y) + ['z']*len(z)
})

# Perform Dunn's post-hoc test
dunn_result = sp.posthoc_dunn(data, val_col='value', group_col='group', p_adjust='bonferroni')

print("Dunn's Post-Hoc Test Results (Bonferroni-corrected p-values):")
display(dunn_result)

# Interpret the results
alpha = 0.05
print(f"\nPairwise comparisons where p < {alpha} are statistically significant:")
for i in dunn_result.index:
    for j in dunn_result.columns:
        if i != j and dunn_result.loc[i, j] < alpha:
            print(f"  - Group {i} vs Group {j}: p = {dunn_result.loc[i, j]:.3f} (Significant)")
```
