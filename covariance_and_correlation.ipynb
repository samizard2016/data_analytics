```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

$\sigma_{XY} = \frac{1}{N} \sum_{i=1}^{N} (x_i - \mu_X)(y_i - \mu_Y)$ for population covariance whereas for sample covariance we use the following (when we use the cov method in pandas)

$s_{XY} = \frac{1}{n-1} \sum_{i=1}^{n} (x_i - \overline{x})(y_i - \overline{y})$.

Correlation: $\rho_{XY} = \frac{\text{Cov}(X,Y)}{\sigma_X \sigma_Y}$

$r_{xy} = \frac{\sum_{i=1}^{n} (x_i - \overline{x})(y_i - \overline{y})}{\sqrt{\sum_{i=1}^{n} (x_i - \overline{x})^2 \sum_{i=1}^{n} (y_i - \overline{y})^2}}$

Covaraince can show the direction of the variance across features whereas correlation helps us find the strength of associations across features when scaled it between 0 and 1 with sign (direction of movement).



```python
data = sns.load_dataset("iris")
x = data.iloc[:,:-1]
cov = x.cov()
corr = x.corr()
cov, corr
```

**The three main correlation methods available in pandas' corr() method are:**

**pearson (default)**: This is the standard Pearson correlation coefficient, which measures the linear relationship between two datasets. It returns a value between -1 and 1, where 1 indicates a perfect positive linear correlation, -1 indicates a perfect negative linear correlation, and 0 indicates no linear correlation.

**kendall**: This is Kendall's Tau correlation coefficient. It's a non-parametric measure of the strength and direction of association that exists between two ranked variables. It's less sensitive to outliers and works well with ordinal data.

**spearman**: This is Spearman's Rank correlation coefficient. Like Kendall's Tau, it's a non-parametric measure that assesses how well the relationship between two variables can be described using a monotonic function. It's also robust to outliers and can detect non-linear monotonic relationships.
