```python
import numpy as np
import pandas as pd

# Group A: polarized (extremes)
group_A = np.array([1]*25 + [5]*25)

# Group B: neutral (middle-heavy)
group_B = np.array([3]*50)

# Put into DataFrame
df = pd.DataFrame({
    'score': np.concatenate([group_A, group_B]),
    'group': ['A']*50 + ['B']*50
})

# Quick check
print(df.groupby('group')['score'].mean())
```

```python
from scipy.stats import ttest_ind

t_stat, p_val = ttest_ind(group_A, group_B)
print(p_val)
```

```python
from scipy.stats import mannwhitneyu

u_stat, p_val = mannwhitneyu(group_A, group_B, alternative='two-sided')
print(p_val)
```

```python
import statsmodels.api as sm
from statsmodels.miscmodels.ordinal_model import OrderedModel

# Encode group (A=1, B=0)
df['group_code'] = (df['group'] == 'A').astype(int)

model = OrderedModel(df['score'],
                     df[['group_code']],
                     distr='logit')

res = model.fit(method='bfgs')
print(res.summary())
```

```python
from scipy.stats import chi2_contingency

# Contingency table
table = pd.crosstab(df['group'], df['score'])

chi2, p, dof, exp = chi2_contingency(table)
print(p)
```
